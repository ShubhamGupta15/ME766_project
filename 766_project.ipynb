```python
# Importing Image and ImageOps module from PIL package
from PIL import Image, ImageOps
import os, sys

# creating an og_image object
for y in range(1,10):
    for x in range(1,6):
        x = str(x)
        y = str(y)
        og_image = Image.open("C:/Users/Anish/Downloads/GTdb_crop/cropped_faces/s0" + y + "_0" + x + ".jpg")
        #og_image.show()

        # applying grayscale method
        gray_image = ImageOps.grayscale(og_image)
        #gray_image.show()

        im1 = gray_image.save("s0" + y + "_0" + x  + ".jpg")

        im = Image.open("C:/Users/Anish/s0" + y + "_0" + x + ".jpg", 'r')
        pix_val = list(im.getdata())
        #len(pix_val)
        
        
        with open("s0" + y + "_0" + x  + ".txt", 'w') as f:
            for item in pix_val:
                f.write("%s\n" % item)

path = "C:/Users/Anish/New Folder/Grayscale_images/"
dirs = os.listdir( path )
def resize():
    for item in dirs:
        if os.path.isfile(path+item):
            im = Image.open(path+item)
            f, e = os.path.splitext(path+item)
            imResize = im.resize((150,200), Image.ANTIALIAS)
            imResize.save(f + ' resized.jpg', 'JPEG', quality=90)

resize()

for y in range(1,10):
    for x in range(1,6):
        x = str(x)
        y = str(y)
        im = Image.open("C:/Users/Anish/New folder/Rescaled_Images/s0" + y + "_0" + x + " resized.jpg", 'r')
        pix_val = list(im.getdata())
        #len(pix_val)
        
        
        with open("s0" + y + "_0" + x  + "resized.txt", 'w') as f:
            for item in pix_val:
                f.write("%s\n" % item)
                
#file1 = open('s01_01.txt', 'r')
#print(file1.read())
#file1.close()
```
